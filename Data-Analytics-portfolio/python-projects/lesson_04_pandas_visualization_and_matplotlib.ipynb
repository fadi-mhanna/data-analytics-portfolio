```python
import pandas as pd

sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [500, 700, 650, 900, 800]
}

df = pd.DataFrame(sales)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['Day'], df['Revenue'])
plt.title('Revenue trend')
plt.xlabel('Day')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(df['Day'], df['Revenue'])
plt.title('Revenue trend')
plt.xlabel('Day')
plt.ylabel('Revenue')
plt.show()
```

```python
sales = {
    'Category': ['Beer', 'Food', 'Cocktails', 'Wine'],
    'Revenue': [4500, 6200, 3800, 2100]
}

df = pd.DataFrame(sales)
```

```python
plt.bar(df['Category'], df['Revenue'])
plt.title('Revenue by category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(df['Category'], df['Revenue'])
plt.title('Revenue by category')
plt.xlabel('Category')
plt.ylabel('Revenue')   
plt.show()
```

```python
plt.pie(df['Revenue'], labels=df['Category'], autopct = '%1.1f%%')
plt.title('Revenue distribution by category')
plt.show()
```

```python
sales = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Revenue': [5000, 5500, 5200, 7000, 6800]
}

df = pd.DataFrame(sales)
```

```python
plt.plot(df['Month'], df['Revenue'])
plt.title('Monthly revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```
