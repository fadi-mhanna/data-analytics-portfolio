```python
import pandas as pd
```

```python
sales = {
    "Day": [
        "Monday",
        "Monday",
        "Tuesday",
        "Tuesday",
        "Wednesday",
        "Wednesday"
    ],

    "Category": [
        "Food",
        "Drinks",
        "Food",
        "Drinks",
        "Food",
        "Drinks"
    ],

    "Revenue": [
        500,
        700,
        1000,
        600,
        1500,
        300
    ]
}

df = pd.DataFrame(sales)
```

```python
df
```

```python
df.pivot_table(
    values="Revenue",
    index="Day",
    columns="Category",
    aggfunc="sum"
)
```

```python
df.pivot_table(
    values="Revenue",
    index="Day",
    columns="Category",
    aggfunc="mean"
)
```
