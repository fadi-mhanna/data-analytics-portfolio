```python
import pandas as pd
sales = {
    "Day": [
        "Monday",
        "Monday",
        "Tuesday",
        "Tuesday",
        "Wednesday",
        "Wednesday"
    ],

    "Revenue": [
        500,
        700,
        1000,
        600,
        1500,
        300
    ]
}

df = pd.DataFrame(sales)

df
```

```python
df.groupby(df['Day']).sum()
```

```python
df.groupby(df['Day']).mean()
```

```python
df.groupby(df['Day']).mean().idxmax()
```

```python
df.groupby(df['Day'])['Revenue'].count()
```

```python
df.groupby('Day')['Revenue'].agg(['sum', 'mean', 'count'])
```

```python
df.groupby('Day')['Revenue'].agg(['sum', 'mean', 'max', 'min', 'count'])
```

```python
sales = {
    "Day": [
        "Monday",
        "Monday",
        "Tuesday",
        "Tuesday",
        "Wednesday",
        "Wednesday"
    ],

    "Category": [
        "Food",
        "Drinks",
        "Food",
        "Drinks",
        "Food",
        "Drinks"
    ],

    "Revenue": [
        500,
        700,
        1000,
        600,
        1500,
        300
    ]
}

df = pd.DataFrame(sales)

df
```

```python
df.groupby(['Day', 'Category'])['Revenue'].sum()
```

```python
df.groupby(['Day', 'Category'])['Revenue'].mean()
```

```python
df.groupby('Category')['Revenue'].max()
```
