```python
import pandas as pd
```

```python
sales = { 'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
           'Revenue': [1200, 1500, 1800, 2200, 3500] }
```

```python
df= pd.DataFrame(sales)
df
```

```python
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
maximum_revenue = df['Revenue'].max()
print("Maximum Revenue:", maximum_revenue)
```

```python
the_mean_revenue = df['Revenue'].mean()
print("Mean Revenue:", the_mean_revenue)
```

```python
df['Day'][df['Revenue'] == maximum_revenue]
```
