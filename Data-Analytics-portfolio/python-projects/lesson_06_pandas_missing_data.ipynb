```python
import pandas as pd

data = {
    'Employee': ['Fadi', 'Sean', 'John', 'Sara'],
    'Revenue': [500, None, 700, None]
}

df = pd.DataFrame(data)

df
```

```python
df.isna().sum()
```

```python
df['Revenue'] = df['Revenue'].fillna(0)
```
